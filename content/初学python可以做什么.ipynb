### 分析股票数据，指导投资

#### 用tushare获取某公司交易行情数据，分析大单交易

```python
import pandas as pd
import numpy as np
import tushare as ts
import datetime
import hvplot.pandas
#获取某股票某天的大单交易，并查看数据
mydf = ts.get_sina_dd('300113','2018-11-19')
mydf.head()
```

```python
#绘制点状图，查看大单分布情况
mydf.hvplot(x='price', y='volume',kind='scatter',title='test')
```
